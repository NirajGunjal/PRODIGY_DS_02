```python
mean = data1.groupby('Pclass')['Age'].mean()
mean
```

```python
def impute_age(cols):
 Age = cols[0]
 Pclass = cols[1]
 if pd.isnull(Age):
     if Pclass == 1:
         return 38
     elif Pclass == 2:
         return 30
     else:
         return 25
 else:
     return Age
```

```python
data1['Age'] = data1[['Age', 'Pclass']].apply(impute_age, axis=1)
```

```python
new_df = pd.concat([data1, dummies_sex, dummies_embarked], axis=1)
new_df.head()
```
